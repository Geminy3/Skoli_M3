```python
nlp = spacy.load("fr_core_news_sm")
```

```python
def clean_corpus(text):
    text = text.strip()
    text = re.sub(".{4:}", ".", text)
    text = re.sub("\s\w\s\w*", "", text)

    return(text)
```

```python
corpus = []
error = {"file": [], "text" : []}
for file in os.listdir("../export2txt"):
    #print(file)
    with open("export2txt/"+file, "r", encoding="utf-8") as f:
        text = clean_corpus(f.read())
        if re.search("(cid:\d\d)", text):
            error["file"].append(file)
            error["text"].append(text)
        # if re.search("(\w\s\w\s)*}", text):
        #     print(file, text)
        # text = clean_corpus(text)
        # text = nlp(text)
        # for token in text:
        #     print(token)
```
